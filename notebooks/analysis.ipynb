```python
import pandas as pd

orders = pd.read_csv("data/orders.csv")
orders.head()
```

```python
import pandas as pd

kpi = pd.Series({
    "Total Orders": len(orders),
    "Unique Customers": orders["customer_id"].nunique(),
    "Total Revenue (CNY)": orders["sell_price_cny"].sum(),
    "Total Profit (CNY)": orders["profit_cny"].sum(),
    "Avg Margin %": orders["margin_pct"].mean(),
    "On-time Rate": (orders["delay_days"] == 0).mean(),
    "Avg Delivery Days": orders["actual_days"].mean(),
    "Complaint Rate": orders["complaint_flag"].mean(),
    "Refund Rate": orders["refund_flag"].mean(),
    "Avg Sentiment (1-5)": orders["sentiment_score"].mean(),
})

kpi.to_frame("value")
```

```python
kpi_round = kpi.copy()

kpi_round["Total Revenue (CNY)"] = round(kpi_round["Total Revenue (CNY)"], 0)
kpi_round["Total Profit (CNY)"] = round(kpi_round["Total Profit (CNY)"], 0)
kpi_round["Avg Margin %"] = round(kpi_round["Avg Margin %"] * 100, 2)
kpi_round["On-time Rate"] = round(kpi_round["On-time Rate"] * 100, 2)
kpi_round["Complaint Rate"] = round(kpi_round["Complaint Rate"] * 100, 2)
kpi_round["Refund Rate"] = round(kpi_round["Refund Rate"] * 100, 2)

kpi_round.to_frame("value")
```

```python
import numpy as np

lane = orders[(orders["market"]=="US") & (orders["shipping_mode"]=="Standard")].copy()

# 原始平均margin
base_margin = lane["margin_pct"].mean()

# 升级后：运费 +15%
ship_up = lane["shipping_cost_cny"] * 1.15

# 其他成本不变
other_cost = (
    lane["cogs_cny"] +
    lane["warehouse_fee_cny"] +
    lane["warehouse_fee_cny"]*0 +  # 占位，不影响
    lane["coupon_value_cny"] +
    lane["gift_cost_cny"]
)

# 原始退款金额
refund_base = lane["refund_amount_cny"].copy()

# 只对 15+ days 的退款金额做缩放：乘 (1 - r)
mask_extreme = lane["delay_bucket"] == "15+ days"

def sim_margin(r):
    refund_sim = refund_base.copy()
    refund_sim.loc[mask_extreme] = refund_sim.loc[mask_extreme] * (1 - r)
    profit_sim = lane["sell_price_cny"] - (other_cost + ship_up + refund_sim)
    return (profit_sim / lane["sell_price_cny"]).mean()

# 扫描 r=0%..100%
rs = np.linspace(0, 1, 21)
sim_margins = [sim_margin(r) for r in rs]

# 找到 break-even：模拟margin >= 原始margin 的最小r
breakeven = None
for r, m in zip(rs, sim_margins):
    if m >= base_margin:
        breakeven = r
        break

base_margin, breakeven, list(zip(rs, sim_margins))[:5]
```

```python
import numpy as np

lane = orders[(orders["market"]=="AU") & (orders["shipping_mode"]=="Standard")].copy()

base_margin = lane["margin_pct"].mean()
ship_up = lane["shipping_cost_cny"] * 1.15

other_cost = (
    lane["cogs_cny"] +
    lane["warehouse_fee_cny"] +
    lane["coupon_value_cny"] +
    lane["gift_cost_cny"]
)

refund_base = lane["refund_amount_cny"].copy()
mask_extreme = lane["delay_bucket"] == "15+ days"

def sim_margin(r):
    refund_sim = refund_base.copy()
    refund_sim.loc[mask_extreme] = refund_sim.loc[mask_extreme] * (1 - r)
    profit_sim = lane["sell_price_cny"] - (other_cost + ship_up + refund_sim)
    return (profit_sim / lane["sell_price_cny"]).mean()

rs = np.linspace(0, 1, 21)
sim_margins = [sim_margin(r) for r in rs]

breakeven = None
for r, m in zip(rs, sim_margins):
    if m >= base_margin:
        breakeven = r
        break

base_margin, breakeven, list(zip(rs, sim_margins))[:5]
```
